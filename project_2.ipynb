```python
# slecting the feature we need for recomndation
selected_features = ['Genre_1','Genre_2','Genre_3','Genre_4','Genre_5','Genre_6','keywords','tagline','cast','director','original_language']
print(selected_features)
```

```python
# seeing all the filled value
display (data.isna().sum())
```

```python
display (data[selected_features].head())
```

```python
# seeing for all null value
display (data[selected_features].isna().sum())
```

```python
for feature in selected_features:
  data[feature] = data[feature].fillna('')
display (data.head())
```

```python
#comining data before we use in bag of words
combined_features = data['Genre_1'] +' ' + data['Genre_2'] +' '+ data['Genre_3'] +' '+ data['Genre_4'] +' '+ data['Genre_5']+' '+data['Genre_6']+'  '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']+' '+data['original_language']
display (combined_features)
```

```python
# we are fiting data for bag of words
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
display (feature_vectors.shape)
print (feature_vectors)
```

```python
# using bag of words to make array to see which is have common with each other
similarity = cosine_similarity(feature_vectors)
print  (similarity )
```

```python
display(similarity.shape)
```

```python
# Taking movie name ffrom user
movie_name = input(' Enter your favourite movie name : ')
```

```python
# All the movies
list_of_all_titles = data['title'].tolist()
print(list_of_all_titles)
```

```python
len(list_of_all_titles)
```

```python
# colsest movies name to one selected by users
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
# most closest to name given by user
close_match = find_close_match[0]
print(close_match)
```

```python
# index no of movie
index_of_the_movie = data[data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
# The row of array of index from bag of words
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
# sorteing the row so we can get the movies closest to the number for recommandation
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
# showing the top 30 movie for recommandation
print('Movies suggested for you : \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = data[data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
